```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def J(theta):
    return (theta - 2.5) ** 2 -1
def dJ(theta):
    return 2 * (theta - 2.5)
```

```python
x = np.linspace(0,5,50)
y = J(x)
```

```python
theta = 0
eta = 0.1
eplison = 1e-8
theta_history = []
iter = 0
iter_num = 1000
while(iter < iter_num):
    last_theta = theta
    gradient = dJ(theta)
    
    theta = theta - eta * gradient
    theta_history.append(theta)
    if np.absolute(J(theta) - J(last_theta)) < eplison:
        break
    iter += 1
    
y_history = [J(theta) for theta in theta_history]

plt.plot(x,y,color = "green")
plt.plot(theta_history,y_history,color = "red",marker = "+")
```

```python
def gradient(init_theta,iter_num = 1e4,eta = 0.1,eplison = 1e-8):
    theta = init_theta
    iter = 0
    theta_history =[theta]
    while(iter <  iter_num):
        last_theta = theta
        theta = theta - eta * dJ(theta)
        theta_history.append(theta)
        if np.absolute(J(theta) - J(last_theta)) < eplison:
            break
        iter += 1
    return theta_history
def plot_theta_history(base_theta,theta_history):
    plt.plot(base_theta,[J(theta) for theta in base_theta])
    plt.plot(theta_history,[J(theta) for theta in theta_history],color = "red",marker = "*")
base_theata = np.linspace(0,5,100)
theta_history = gradient(init_theta= 0)
```

```python
plot_theta_history(base_theta=base_theata,theta_history= theta_history)
```
