```python
boston = datasets.load_boston()
```

```python
X = boston.data
y = boston.target
```

```python
X.shape,y.shape
```

```python
x_lim = X[:,5]
x_lim.shape,y.shape
```

```python
x_lim = x_lim[y< 50]
y = y[y<50]
```

```python
x_lim.shape,y.shape
```

```python
plt.scatter(x_lim,y)
```

```python
from SimpleLinearRegression import SimpleLinearRegression2
```

```python
srf = SimpleLinearRegression2()
srf.fit(x_lim,y)
y_predict = srf.predict(x_lim)
```

```python
plt.scatter(x_lim,y)
plt.scatter(x_lim,y_predict,color = "red")
```

### MSE

```python
mse = ((y - y_predict).dot(y - y_predict)) / len(y)
mse
```

### RMSE

```python
rmse = np.sqrt(mse)
```

```python
rmse
```

### MAE

```python
mae = np.sum(np.absolute(y - y_predict)) / len(y)
```

```python
mae
```

### R2

```python
R2 = 1- (y - y_predict).dot(y - y_predict) / (np.mean(y) - y).dot(np.mean(y) - y)
R2
```

```python
R2 = 1 - mse / np.var(y)
R2
```
