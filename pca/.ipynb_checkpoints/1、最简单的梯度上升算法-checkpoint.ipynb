```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-20,100,200)
y = (x -1) * (x -10) * (x -100)
```

```python
plt.plot(x,y)
```

```python
def J(theta):
    return (theta -1) * (theta -10) * (theta -100)
def dJ(theta):
    return (3 * theta ** 2) - (222 *theta) + 1110
```

```python
iter_num =1000
eta = 0.001
eplison = 1e-8

iter = 0
init_theta = -15
theta = init_theta
theta_history = []
while(iter < iter_num):
    theta_history.append(theta)
    last_theta = theta
    gradient = dJ(theta)
    theta = theta + eta * gradient
    if np.abs(J(theta) - J(last_theta)) < eplison:
        break
    iter += 1

theta_history = np.array(theta_history)
y_history = J(theta_history)

plt.plot(x,y)
plt.plot(theta_history,y_history,color = "red",marker = "+")
```
