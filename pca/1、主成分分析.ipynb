```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X[:,0] = np.random.uniform(0,100,size=100)
X[:,1] = 0.75 * X[:,0] + 3 + np.random.normal(0,10,len(X[:,]))
```

```python
plt.scatter(X[:,0],X[:,1])
```

```python
def demean(X):
    return X - np.mean(X)
def f(w,X):
    return np.sum(X.dot(w) ** 2) * 2 / len(X)
def df(w,X):
    return X.T.dot(X.dot(w))
def direction(w):
    return w / np.linalg.norm(w)
```

```python
def gradient_ascent(init_w,X,iter_num = 1e4,eta = 0.001,eplison = 1e-8):
    w = direction(init_w)
    iter = 0
    while(iter < iter_num):
        last_w = w
        gradient = df(w,X)
        w = direction(w + eta * gradient)
        if (np.abs(f(w,X) - f(last_w,X))) < eplison :
            break
        iter += 1
    return w
X = demean(X)
init_w = np.random.normal(size=X.shape[1])  
final_w = gradient_ascent(init_w,X)
final_w
```

```python
X_trans = np.empty((X.shape))
```

```python
X_trans.shape
```

```python
X_trans[:,0] = X[:,0]
```

```python
X_trans[:,1] = X_trans[:,0] + final_w[0] + final_w[1]
```

```python
plt.scatter(X[:,0],X[:,1])
plt.scatter(X_trans[:,0],X_trans[:,1],color = "red",marker = "*")
```
