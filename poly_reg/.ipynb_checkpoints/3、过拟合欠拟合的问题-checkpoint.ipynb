```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler,PolynomialFeatures
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
x = np.random.uniform(-3,3,100)
X = x.reshape(-1,1)
y = 0.5 * x ** 2 + x + 2 + np.random.normal(0,1,size=len(x))
plt.scatter(x,y)
```

```python
lin_reg = LinearRegression()
lin_reg.fit(X,y)
y_predict = lin_reg.predict(X)
plt.scatter(x,y)
plt.plot(x,y_predict)
```

```python
mean_squared_error(y,y_predict)
```

```python
def polynomial_regression(degree):
    return Pipeline([
        ('polyfeature',PolynomialFeatures(degree= degree)),
        ('standscaler',StandardScaler()),
        ('reg',LinearRegression())
    ])
```

```python
poly_reg2 = polynomial_regression(2)
poly_reg2.fit(X,y)
y_predict2 = poly_reg2.predict(X)
plt.scatter(x,y)
# plt.scatter(x,y_predict2)

plt.plot(x[np.argsort(x)],y_predict2[np.argsort(x)],color = "red")
```

```python
mean_squared_error(y,y_predict2)
```

```python
poly_reg10 = polynomial_regression(10)
poly_reg10.fit(X,y)
y_predict10 = poly_reg10.predict(X)

plt.scatter(x,y)
plt.plot(x[np.argsort(x)],y_predict10[np.argsort(x)],color = "red")
```

```python
poly_reg100 = polynomial_regression(100)
poly_reg100.fit(X,y)
y_predict100 = poly_reg100.predict(X)

plt.scatter(x,y)
plt.plot(x[np.argsort(x)],y_predict100[np.argsort(x)],color = "red")
```

```python
mean_squared_error(y,y_predict100)
```
